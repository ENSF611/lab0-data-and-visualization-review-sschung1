```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('auto-mpg.data.csv', na_values='?')
```

```python
data.plot()
```

```python
data['mpg'].plot.hist()
```

```python
data.origin.value_counts()
```

```python
axs = data.hist(column='mpg', by='origin')
```

```python
axs = data.hist(column='mpg', by='origin')
axs[0, 0].set(title='USA', ylim=[0, 50])
axs[0, 1].set(title='Europe', ylim=[0,50])
axs[1, 0].set(title='Japan', ylim=[0, 50])
```

```python
data.plot.scatter('mpg', 'weight', c='origin', colormap='winter')
```

```python
fig, ax = plt.subplots()
data.plot.scatter('mpg', 'weight', c='origin', colormap='viridis', ax=ax)
```

```python
colors = {1: 'tab:blue', 2: 'tab:orange', 3: 'tab:purple'}
fig, ax = plt.subplots()
for key, group in data.groupby(by='origin'):
    group.plot.scatter('mpg', 'weight', c=colors[key], label=key, ax=ax)
```

```python
import seaborn as sns
```

```python
sns.displot(x='mpg', col='origin', data=data)
```

```python
sns.displot(x='mpg', hue='origin', data=data)
```

```python
sns.displot(x='mpg', hue='origin', data=data, multiple='stack')
```

```python
sns.boxplot(x='origin', y='mpg', data=data)
```

```python
data['origin_str'] = data['origin'].replace([1, 2, 3], ['USA', 'Europe', 'Japan'])
ax = sns.scatterplot(x='mpg', y='weight', data=data, hue='origin_str')
```

```python
ax = sns.lmplot(x='mpg', y='weight', data=data, hue='origin_str')
```

```python
sns.pairplot(data, vars=['mpg', 'weight', 'horsepower', 'cylinders'], hue='origin_str')
```

```python
g = sns.heatmap(data[['mpg', 'weight', 'horsepower', 'cylinders']].corr(method='spearman'),
               annot=True)
```
